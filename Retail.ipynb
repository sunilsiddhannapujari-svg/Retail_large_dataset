```python
# IQR Method – Rule-based detection

Q1 = Retail['product_price'].quantile(0.25)
Q3 = Retail['product_price'].quantile(0.75)

IQR = Q3 - Q1

lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR

outliers_price = Retail[
    (Retail['product_price'] < lower) |
    (Retail['product_price'] > upper)
]

outliers_price.head()

#This will correctly display the first few rows of detected outliers.

# If you still get an empty DataFrame, it means your dataset has no values outside the IQR thresholds.
```
